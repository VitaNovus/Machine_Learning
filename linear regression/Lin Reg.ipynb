# Linear Regression

```python
# importing the essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Student_Performance.csv')
df
```

```python
# dropping the unessential columns
df.drop('Extracurricular Activities', axis= 1, inplace = True)
df
```

```python
# checking the correlation
df.corr()
```

```python
# differentiating the features and output
x = df[['Hours Studied','Previous Scores','Sleep Hours','Sample Question Papers Practiced']]
y = df['Performance Index']
```

```python
# spliting the dataset into training and testing datasets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42)
```

```python
# creating the linear regression machine
from sklearn.linear_model import LinearRegression
reg = LinearRegression(n_jobs = -1)
```

```python
# fitting the training dataset
reg.fit(x_train, y_train)
```

```python
# predicting for the test dataset
y_pred = reg.predict(x_test)
```

```python
# checking the metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
print("mse: ", mean_squared_error(y_test, y_pred))
print("mae: ", mean_absolute_error(y_test, y_pred))
print("r2_score: ", r2_score(y_test, y_pred))
```
